```python
## Libraries Used.
from astroquery.gaia import Gaia
import matplotlib.pyplot as plt
import math
```

```python
def extract_dl_ind(datalink_dict, key, figsize = [15,5], fontsize = 12, linewidth = 2, show_legend = True, show_grid = True):
    ""
    "Extract individual DataLink products and export them to an Astropy Table"
    ""
    dl_out  = datalink_dict[key][0].to_table()
    if 'time' in dl_out.keys():
        plot_e_phot(dl_out, colours  = ['green', 'red', 'blue'], title = 'Epoch photometry', fontsize = fontsize, show_legend = show_legend, show_grid = show_grid, figsize = figsize)
    if 'wavelength' in dl_out.keys():
        if len(dl_out) == 343:  title = 'XP Sampled'
        if len(dl_out) == 2401: title = 'RVS'
        plot_sampled_spec(dl_out, color = 'blue', title = title, fontsize = fontsize, show_legend = False, show_grid = show_grid, linewidth = linewidth, legend = '', figsize = figsize)
    return dl_out


def plot_e_phot(inp_table, colours  = ['green', 'red', 'blue'], title = 'Epoch photometry', fontsize = 12, show_legend = True, show_grid = True, figsize = [15,5]):
    ""
    "Epoch photometry plotter. 'inp_table' MUST be an Astropy-table object."
    ""
    fig      = plt.figure(figsize=figsize)
    xlabel   = f'JD date [{inp_table["time"].unit}]'
    ylabel   = f'magnitude [{inp_table["mag"].unit}]'
    gbands   = ['G', 'RP', 'BP']
    colours  = iter(colours)

    plt.gca().invert_yaxis()
    for band in gbands:
        phot_set = inp_table[inp_table['band'] == band]
        plt.plot(phot_set['time'], phot_set['mag'], 'o', label = band, color = next(colours))
    make_canvas(title = title, xlabel = xlabel, ylabel = ylabel, fontsize= fontsize, show_legend=show_legend, show_grid = show_grid)
    plt.show()


def plot_sampled_spec(inp_table, color = 'blue', title = '', fontsize = 14, show_legend = True, show_grid = True, linewidth = 2, legend = '', figsize = [12,4], show_plot = True):
    ""
    "RVS & XP sampled spectrum plotter. 'inp_table' MUST be an Astropy-table object."
    ""
    if show_plot:
        fig      = plt.figure(figsize=figsize)
    xlabel   = f'Wavelength [{inp_table["wavelength"].unit}]'
    ylabel   = f'Flux [{inp_table["flux"].unit}]'
    plt.plot(inp_table['wavelength'], inp_table['flux'], '-', linewidth = linewidth, label = legend)
    make_canvas(title = title, xlabel = xlabel, ylabel = ylabel, fontsize= fontsize, show_legend=show_legend, show_grid = show_grid)
    if show_plot:
        plt.show()


def make_canvas(title = '', xlabel = '', ylabel = '', show_grid = False, show_legend = False, fontsize = 12):
    ""
    "Create generic canvas for plots"
    ""
    plt.title(title,    fontsize = fontsize)
    plt.xlabel(xlabel,  fontsize = fontsize)
    plt.ylabel(ylabel , fontsize = fontsize)
    plt.xticks(fontsize = fontsize)
    plt.yticks(fontsize = fontsize)
    if show_grid:
        plt.grid()
    if show_legend:
        plt.legend(fontsize = fontsize*0.75)
```

```python
query = f"SELECT source_id FROM gaiadr3.vari_eclipsing_binary"

job     = Gaia.launch_job_async(query)
results = job.get_results()
print(f'Table size (rows): {len(results)}')
```

```python
total_ids = len(results['source_id'].value.data)
batch = 5000
ids = results['source_id'].value.data
```

```python
retrieval_type = 'EPOCH_PHOTOMETRY'          # Options are: 'EPOCH_PHOTOMETRY', 'MCMC_GSPPHOT', 'MCMC_MSC', 'XP_SAMPLED', 'XP_CONTINUOUS', 'RVS', 'ALL'
data_structure = 'INDIVIDUAL'   # Options are: 'INDIVIDUAL', 'COMBINED', 'RAW'
data_release   = 'Gaia DR3'     # Options are: 'Gaia DR3' (default), 'Gaia DR2'

k=0
for i in range(0,math.ceil(total_ids/batch)):
    if i * batch > total_ids:
        batch_ids = ids[i*batch:total_ids]
    else:
        batch_ids = ids[i*batch:(i+1)*batch]
    datalink = Gaia.load_data(ids=batch_ids, data_release = data_release, retrieval_type=retrieval_type, data_structure = data_structure, verbose = False, output_file = None)
    dl_keys  = [inp for inp in datalink.keys()]
    dl_keys.sort()
    for dl_key in dl_keys:
        datalink[dl_key][0].to_table().to_pandas().to_csv("./vari_eclipsing_binary/"+dl_key+".csv")
    k+=batch
    print(k," downloaded")

print()
print(f'The following Datalink products have been downloaded:')
for dl_key in keys:
    print(f' * {dl_key}')
```

```python
def convert_to_list(string):
    return [float(i) if i != '' else 0 for i in string.split(', ')]
```

```python
def process_dataset(df):
    columns = ['time', 'mag', 'flux']
    for i in columns:
        df[i] = df[i].apply(lambda x: x[1:-1])
        df = df[~df[i].isnull()]
        df[i] = df[i].apply(convert_to_list)
    return df
```

```python
df_lpv = process_dataset(df_lpv)
df_cep = process_dataset(df_cep)
df_ecl = process_dataset(df_ecl)
df_rr = process_dataset(df_rr)
```

```python
def plot_records(df,variable,name,band):
    if band == 'G':
        indexes = [0,3,6,9]
    if band == 'RP':
        indexes = [1,4,7,10]
    if band == 'BP':
        indexes = [2,5,8,11]
    fig, ax = plt.subplots(2,2,figsize = (20,8))
    ax[0][0].plot(df['time'][indexes[0]], df[variable][indexes[0]], color = band[0].lower())
    ax[0][0].scatter(df['time'][indexes[0]], df[variable][indexes[0]], color = 'black')
    ax[0][0].set_title('Record 1')
    ax[0][1].plot(df['time'][indexes[1]], df[variable][indexes[1]], color = 'black')
    ax[0][1].scatter(df['time'][indexes[1]], df[variable][indexes[1]], color = band[0].lower())
    ax[0][1].set_title('Record 2')
    ax[1][0].plot(df['time'][indexes[2]], df[variable][indexes[2]], color = band[0].lower())
    ax[1][0].scatter(df['time'][indexes[2]], df[variable][indexes[2]], color = 'black')
    ax[1][0].set_title('Record 3')
    ax[1][1].plot(df['time'][indexes[3]], df[variable][indexes[3]], color = band[0].lower())
    ax[1][1].scatter(df['time'][indexes[3]], df[variable][indexes[3]], color = 'black')
    ax[1][1].set_title('Record 4')
    plt.suptitle(f'Visualizing {variable} of class {name} of band {band}')
    for i in ax.flatten():
        i.set_xlabel('Time')
        i.set_ylabel(f'{variable}')
        i.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plot_records(df_lpv,'mag','LPV',"G")
plot_records(df_cep,'mag','CEP',"G")
plot_records(df_rr,'mag','RR',"G")
plot_records(df_ecl,'mag','ECL',"G")
```

```python
def get_statistics(df, band, variable):
    temp =df[df['band'] == band] 
    arr_temp = np.hstack(temp[variable])
    temp_df = pd.DataFrame(arr_temp, columns = [variable])
    return temp_df.describe()
```

```python
def print_statistics(df,variable,name,band):
    temp = get_statistics(df,band,variable)
    pt = PrettyTable()
    pt.field_names = ['Statistic','Value']
    print(f'Statistics of {variable} for class {name}.')
    for i in range(len(temp)):
        pt.add_row([temp.index[i],round(temp[variable][temp.index[i]],4)])
    print(pt)
```

        <li> Transform y from classes to numbers to feed into the model.</li>
        <li> Perform a tri-split to have train, validation and test data.</li>
        <li> Pre-pad the X columns to maximum length to solve any randomness in lengths</li>
        <li> Convert tensorflow datasets for train_data, validation_data and test_data.</li>
    </ol>
</font>

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv1D, Dense, Flatten, Embedding, SimpleRNN, Input, Reshape, GRU, LSTM
from tensorflow.keras.preprocessing import sequence
from keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras import Model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pickle
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay
AUTOTUNE = tf.data.experimental.AUTOTUNE
tf.config.run_functions_eagerly(True)
```

```python
def convert_to_list(string):
    return [float(i) if i != '' else 0 for i in string.split(', ')]

def process_dataset(df):
    columns = ['time', 'mag', 'flux']
    for i in columns:
        df[i] = df[i].apply(lambda x: x[1:-1])
        df = df[~df[i].isnull()]
        df[i] = df[i].apply(convert_to_list)
    return df
```

```python
df = pd.read_csv(r'Final Dataset/final_df_g_complete.csv')
df = process_dataset(df)
df.reset_index(inplace=True)
df.drop(['index'], axis=1, inplace=True)
df.head(2)
```

```python
X = df['mag']
y = df[['label']]
```

```python
def onehot_encoder(y):
    k = len(np.unique(y))
    return np.eye(k)[y]
```

```python
le = LabelEncoder().fit(y['label'])
y = le.transform(y['label'])
y = onehot_encoder(y)
```

```python
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size = 0.1, random_state = 42)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size = 0.2, random_state = 42)
```

```python
max_words = 261
```

```python
prepad_X_train = sequence.pad_sequences(X_train, padding='pre', maxlen = max_words, dtype='float')
prepad_X_train = np.expand_dims(prepad_X_train,axis=2)
prepad_X_val = sequence.pad_sequences(X_val, padding='pre', maxlen = max_words, dtype='float')
prepad_X_val = np.expand_dims(prepad_X_val,axis=2)
prepad_X_test = sequence.pad_sequences(X_test, padding='pre', maxlen = max_words, dtype='float')
prepad_X_test = np.expand_dims(prepad_X_test,axis=2)
```

```python
train_data = tf.data.Dataset.from_tensor_slices((prepad_X_train,y_train))
train_data = train_data.shuffle(buffer_size=len(prepad_X_train))
train_data = train_data.batch(32)
train_data = train_data.prefetch(AUTOTUNE)
```

```python
val_data = tf.data.Dataset.from_tensor_slices((prepad_X_val,y_val))
val_data = val_data.batch(32)
val_data = val_data.prefetch(AUTOTUNE)
```

```python
test_data = tf.data.Dataset.from_tensor_slices((prepad_X_test,y_test))
test_data = test_data.batch(32)
test_data = test_data.prefetch(AUTOTUNE)
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv1D, Dense, Flatten, Embedding, SimpleRNN, Input, Reshape, LSTM
from tensorflow.keras.preprocessing import sequence
from keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras import Model
from tensorflow.keras.models import save_model, load_model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from scipy.interpolate import interp1d
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
import pickle
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay
AUTOTUNE = tf.data.experimental.AUTOTUNE
```

```python
def convert_to_list(string):
    return [float(i) if i != '' else 0 for i in string.split(', ')]

def process_dataset(df):
    columns = ['time', 'mag', 'flux']
    for i in columns:
        df[i] = df[i].apply(lambda x: x[1:-1])
        df = df[~df[i].isnull()]
        df[i] = df[i].apply(convert_to_list)
    return df

def interpolate_(data, num = 261):
    x = data['time']
    y = data['mag']
    new_mag = []
    new_time = []
    for i in range(len(data)):
        t = data.iloc[i]['time']
        x = data.iloc[i]['mag']
        newt = np.linspace(min(t), max(t), num, endpoint=True)
        f = interp1d(t, x)
        newx = f(newt)
        new_mag.append(newx)
        new_time.append(newt)
    df_t = pd.DataFrame({"intrpol_time":[i for i in new_time], "intrpol_mag": [i for i in new_mag]})
    return df_t

def scale(data):
    new_mag = []
    for i in range(len(data)):
        mean_ = np.mean(data.iloc[i]['intrpol_mag'])
        std_ = np.std(data.iloc[i]['intrpol_mag'])
        scaled_v = (data.iloc[i]['intrpol_mag'] - mean_) / std_
        new_mag.append(list(scaled_v))
    return [i for i in new_mag]

def onehot_encoder(y):
    k = len(np.unique(y))
    return np.eye(k)[y]
```

```python
df =  pd.read_csv(r'Final Dataset\final_df_g_complete.csv', index_col = 0)
df = process_dataset(df)
df.reset_index(inplace=True)
df.drop(['index'], axis=1, inplace=True)
df.head(2)
```

### LSTM with Interpolation and time difference as input

```python
def difference_time(time_arr):
    diff_list = [0]
    for i in range(1,len(time_arr)):
        x = time_arr[i] - time_arr[i-1]
        diff_list.append(x)
    return diff_list
```

```python
time_diff = []
for i in df.index:
    diff = difference_time(df['intrpol_time'][i])
    time_diff.append(diff)
```

```python
df['time_diff'] = time_diff
del time_diff
```

```python
time = X['time_diff']
mag = X['scaled_mag']
```

```python
X_stack = np.column_stack([time, mag])
```

```python
X_stack.shape
```

```python
X_stacked_final = []
for i in range(len(X_stack[:,0])):
    X_stacked_final.append(np.column_stack([X_stack[:,0][i], X_stack[:,1][i]]))
```

```python
le = LabelEncoder().fit(y['label'])
y = le.transform(y['label'])
y = onehot_encoder(y)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size = 0.2, random_state = 42)
```

```python
train_data = tf.data.Dataset.from_tensor_slices((X_train,y_train))
train_data = train_data.shuffle(buffer_size=len(X_train))
train_data = train_data.batch(32)
train_data = train_data.prefetch(AUTOTUNE)
```

```python
val_data = tf.data.Dataset.from_tensor_slices((X_val,y_val))
val_data = val_data.batch(32)
val_data = val_data.prefetch(AUTOTUNE)
```

```python
test_data = tf.data.Dataset.from_tensor_slices((X_test,y_test))
test_data = test_data.batch(32)
test_data = test_data.prefetch(AUTOTUNE)
```

```python
input_layer = Input(shape = (261,2))
LSTM1 = LSTM(64, return_sequences=True)(input_layer)
LSTM2 = LSTM(64, return_sequences=True)(LSTM1)
LSTM3 = LSTM(64, return_sequences=False)(LSTM2)
output = Dense(4, activation='softmax')(LSTM3)
lstm_model_4 = Model(inputs=input_layer, outputs=output)
# summarize layers
print(lstm_model_4.summary())
```
